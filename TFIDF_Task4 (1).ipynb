```python
doc1= document_4883[0].to_list()
#print(doc1)
doc2= document_6068[0].to_list()
#print(doc2)
doc3= document_9951[0].to_list()
#print(doc3)
doc4= document_49153[0].to_list()
doc5= document_51816[0].to_list()
doc6= document_63051[0].to_list()
doc7= document_87234[0].to_list()
doc8= document_89904[0].to_list()
doc9= document_95592[0].to_list()
doc10= document_179736[0].to_list()
```

```python
stopwords.words('English')
```

```python
vectorizer=TfidfVectorizer(lowercase=True,stop_words=stopwords.words('English'))
#Learn the vocabulary dictionary and return document-term matrix.
vectors= vectorizer.fit_transform([doc1[0], doc2[0], doc3[0], doc4[0], doc5[0], doc6[0], doc7[0],doc8[0], doc9[0], doc10[0]])

#print(vectors)
```

```python
feature_names=vectorizer.get_feature_names()
 
len(feature_names)
#print(feature_names)
```

```python
dense= vectors.todense()

print(dense)
```

```python
denselist=dense.tolist()
print(denselist)
```

```python
df=pd.DataFrame(denselist,columns=feature_names)
df.index=['UserID_87234', 'UserID_4883', 'UserID_9951', 'UserID_6068', 'UserID_89904', 'UserID_51816', 'UserID_49153', 'UserID_179736', 'UserID_95592', 'UserID_63051']
```
